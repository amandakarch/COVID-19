```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/Users/akarch1/Desktop/Analytics_Portfolio/covid19_cleaned.csv', index_col=0)
print(df.head())
```

```python
states = df.drop(labels='USA', axis=0)
states.sort_index()
```

There have been many speculations about whether COVID-19 will slow down as we move toward warmer months, so I thought I would try to see if there was a relationship between average temperature of states and the number of current COVID-19 US cases, as of April 5, 2020.

```python
sns.relplot(x='Covid_cases_2020_04_05', y='Avg_Temp', data=states, kind='scatter')
```

I decided to remove New York from an additional scatterplot, in order to see if its status as an outlier was strongly affecting the correlation between the two variables.

```python
states_noNY = states.drop(labels='New.York', axis=0)
sns.relplot(x='Covid_cases_2020_04_05', y='Avg_Temp', data=states_noNY, kind='scatter')
```

```python
states[['Covid_cases_2020_04_05', 'Avg_Temp']].corr()
```

The Pearson correlation confirmed what the scatterplots also determined - the current data does not support any correlation between average temperature and COVID-19 cases.

NOTE: Using average temperature data is a way to put different temperatures ranges of states in scale, as it can be higher or lower as we get closer to summer, but it still shows that higher temperatures do not correlate to lower amounts of cases, as has been suggested by media.
